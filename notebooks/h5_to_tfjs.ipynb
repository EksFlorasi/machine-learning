## Convert .h5 model to .json
Using TensorFlow JS.

```python
from keras.models import load_model

FLORA_H5 = "./models/flora/h5/flora-fruits_flowers_vegetables.h5"
flora_model = load_model(FLORA_H5)

FAUNA_H5 = "./models/fauna/h5/fauna-95-89.h5"
fauna_model = load_model(FAUNA_H5)
```

```python
!tensorflowjs_converter --input_format=keras ./models/flora/h5/flora-fruits_flowers_vegetables.h5 ./models/flora/json/flora_model
```

```python
!tensorflowjs_converter --input_format=keras ./models/fauna/h5/fauna-95-89.h5 ./models/fauna/json/fauna_model
```
